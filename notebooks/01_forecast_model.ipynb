```python
import pandas as pd
import matplotlib.pyplot as plt

# Ganti path ke file kamu sendiri
df = pd.read_csv("D:\Latihan ML\sales-forecasting\data\train.csv")

# Lihat 5 baris pertama
print(df.head())

# Cek kolom tanggal
df['date'] = pd.to_datetime(df['date'])

# Agregasi per bulan atau minggu (tergantung dataset)
monthly_sales = df.groupby(pd.Grouper(key='date', freq='M'))['sales'].sum().reset_index()

# Plot tren awal
plt.figure(figsize=(10,5))
plt.plot(monthly_sales['date'], monthly_sales['sales'])
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```
